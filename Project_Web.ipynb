# Web Scraping

**Web scraping del link de Wikipedia donde aparece el listado de películas ganadoras y nominadas cada año en la categoría "Oscar a la mejor película"**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films'

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
soup
```

**Extraemos aquellos resultados que están en el URL sombreados en amarillo, que son las películas ganadoras**

```python
row_winners = soup.find_all('tr', {'style' : 'background:#EEDD82'})
row_winners
```

**Extraemos los elementos de texto de cada línea y hacemos split por cada salto de línea.**

```python
row_winners = [row.text.strip().split("\n") for row in row_winners]
row_winners
```

**Nos queda como resultado una lista de listas con cada título de película, su año, Oscars ganados, y nominaciones**

**Cambiamos el nombre de las columnas para identificarlas mejor.**
